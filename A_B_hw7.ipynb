   - Экспериментальная: пользователи, которым предоставлена возможность оформить подписку на длительный срок.
  
   Запускаем эксперимент продолжительностью 20 дней. Во избежании проблемы подглядывания не меняем количество и состав групп до конца теста и не завершаем его досрочно, особенно в случае появления ожидаемого результата

3. *Тестируемые пользователи* - Пользователи без текущей подписки, совершившие как минимум 2 поездки за последние 30 дней.

4. *Ключевая метрика* - Конверсия пользователей в подписчиков.

5. *Ожидаемый эффект* - Увеличение количества подписчиков среди пользователей в экспериментальной группе на 15%.

6. *План действий в результате эксперимента* - Проводим статистический анализ двух групп, делаем вывод о влиянии внедрения тарифных планов на прирост количества подписчиков. Если наблюдаемый результат соответветствует ожидаемому, то внедряем функцию для всех пользователей. В противном случае - нет или пересматриваем условия и параметры и тестируем снова.
